# Machine Learning 1주차 1개의 입력에서의 선형 회귀
features가 1개 일때 (경사하강법, 손실함수) 1주차 내용 python 구현

## numpy, plot 라이브러리 불러오기

```
import numpy as np
from matplotlib import pyplot as plt
```

## 임의의 훈련 데이터 만들기 및 plot
100개의 입력 X 값과 X 따른 출력 Y 값을 임의로 정해준다
$$
y(x) = x\theta_1 + \theta_0
$$

```
X_train = np.random.rand(100)
theta_0 = 0.2
theta_1 = 0.5
Y = 0.2 * X_train + 0.5
plt.figure(figsize=(8, 6))
plt.scatter(X_train, Y)
plt.show()
```

## Gradient Descent 구현
경사하강법을 이용하여 손실함수 값을 최소화 한다.

### 손실함수
$$
J(\theta_0, \theta_1) = \frac{1}{2m}\sum_{i=1}^m (h_\theta(x_i) - y_i)
$$
모든 항목에 대한 거리의 평균을 낸다. 손실함수 값이 작을 수록 정확도가 올라간다.

### Grident Descent
$$
\theta_j = \theta_j - \alpha\frac{\partial}{\partial \theta_j}J(\theta_0, \theta_1)
$$
편미분을 통하여 각 세타에 맞은 기울기를 구한 후 임의로 정해논 학습률과 곱하여 반영한다. 기울기 값이 0에 수렴 할 수록 정확도가 높아진다. 알파 값이 너무 클 경우 overshooting이 발생한다

```
theta_0_pred = np.random.uniform(-1, 1)
theta_1_pred = np.random.uniform(-1, 1)
learning_rate = 0.7

for epoch in range(151):
  Y_pred = theta_1_pred * X_train + theta_0_pred

  temp0 = theta_0_pred - learning_rate * (Y_pred - Y).mean()
  temp1 = theta_1_pred - learning_rate * ((Y_pred - Y) * X_train).mean()

  theta_0_pred = temp0
  theta_1_pred = temp1

  if epoch % 30 == 0:
    Y_pred = theta_1_pred * X_train + theta_0_pred
    plt.figure(figsize=(8, 6))
    plt.title(str(epoch) + " pred")
    plt.scatter(X_train, Y, label='real')
    plt.scatter(X_train, Y_pred, label='pred')
    plt.legend(loc='best')
    plt.show()
```
